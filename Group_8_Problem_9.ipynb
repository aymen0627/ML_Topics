```python
"""

In this problem, generate 80 unlabeled datapoints using 8 Gaussian distributions. For
every distribution, generate 10 points. The means are centered at (1,0), (0.7,0.7), (0,1), (-0.7,0.7),
(-1,0), (-0.7,-0.7), (0,-1), (0.7,-0.7). For all, the variance is chosen as 0.2. Concatenate the data and
shuffle. Use the Kmean method with the EM algorithm to cluster the data into K = 8 clusters. In one
plot, show the center of each cluster and the label of every datapoint. Repeat with variance chosen as
0.1, 0.4, and 0.7.
"""
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

.2 Variance:

```python
# In this code, we will show how to use K-mean clustering using sklearn...

# Lets first generate some 2-D training data for easier visualization...

# The task is to classify the data inteo K=4 clusters...

# We will use gaussian distributions to generate the data belonging to 4 classes. The means are located at (1,1), (-1,1), (1,-1), (-1,-1). 

# We will use different variances to test our system

# Remember that we have to shuffle the data when we input it to the Kmean algorithm. The algorithm can not know the labels as this an unsupervised learning technique

# Compare the EM to others... see "https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html"

number_of_clusters_K = 8
number_of_data_points_per_cluster = 10
number_of_data_point = number_of_clusters_K*number_of_data_points_per_cluster

data_1_2 = np.random.normal(loc=(1,0), scale = np.sqrt(.2),   size=(number_of_data_points_per_cluster,2)) 
data_2_2 = np.random.normal(loc=(.7,.7), scale = np.sqrt(.2),  size=(number_of_data_points_per_cluster,2)) 
data_3_2 = np.random.normal(loc=(0,1), scale = np.sqrt(.2), size=(number_of_data_points_per_cluster,2)) 
data_4_2 = np.random.normal(loc=(-.7,.7), scale = np.sqrt(.2),  size=(number_of_data_points_per_cluster,2)) 
data_5_2 = np.random.normal(loc=(-1,0), scale = np.sqrt(.2),   size=(number_of_data_points_per_cluster,2)) 
data_6_2 = np.random.normal(loc=(-.7,-.7), scale = np.sqrt(.2),  size=(number_of_data_points_per_cluster,2)) 
data_7_2 = np.random.normal(loc=(0,-1), scale = np.sqrt(.2), size=(number_of_data_points_per_cluster,2)) 
data_8_2 = np.random.normal(loc=(.7,-.7), scale = np.sqrt(.2),  size=(number_of_data_points_per_cluster,2)) 

# concatinate the data and shuffle
data_2 = np.concatenate((data_1_2,data_2_2,data_3_2,data_4_2,data_5_2,data_6_2,data_7_2,data_8_2))

# shuffle the data: 
np.random.shuffle(data_2)
#print(data)
```

```python
# Now, lets call the K mean algorihtm and train using the first 20 data points...

kmeans_2 = KMeans(n_clusters=8, random_state=0, algorithm = "auto").fit(data_2)

# printing the cluster centers: 
#print(kmeans.cluster_centers_)

# plot data points and the learned centers from kmeans...
for (x,y) in data_2: 
  plt.plot(x,y,'bx')
class_cnt = 0
for (x_c,y_c) in kmeans_2.cluster_centers_:
  plt.plot(x_c,y_c,'ro')
  plt.text(x_c,y_c, "Cluster "+str(class_cnt), fontsize=18)
  class_cnt = class_cnt +1

plt.show()
```

```python
# predicted labels for the training data
print(kmeans_2.labels_)

data_to_test = [0,0.1]
pred_label = kmeans_2.predict([data_to_test])
print("datapoint = ",data_to_test,"pred_label", pred_label)

# print the predicted labels: 

for index in range(number_of_data_point):
  plt.plot(data_2[index][0],data_2[index][1], 'bx')
  plt.text(data_2[index][0],data_2[index][1], str(kmeans_2.labels_[index]))
class_cnt = 0
for (x_c,y_c) in kmeans_2.cluster_centers_:
  plt.plot(x_c,y_c,'ro')
  plt.text(x_c,y_c, "Cluster "+str(class_cnt), fontsize=18)
  class_cnt = class_cnt +1
plt.show()
```

A network is given by $h_\theta (x) =w^Tx+b $, where $x,w \in \mathbb{R}^2$ and $b\in \mathbb{R}$. The trainable parameters are $\theta = \{w,b\}$. 

Next, we define the the dataset. 

```python
# Here, we define the intial points of w_1, w_2, and b.
w_1 = 5
w_2 = -4.1
b   = 3

# Define the total loss fucntion f
def function_f(w_1, w_2, b): 
  return 0.25*( (w_1+2*w_2+b-1)**2 + 
                (0.1*w_1 + 3*w_2 + b - 1)**2 +
                (-w_1 - w_2 + b + 1)**2      +
                (-2* w_2 + b + 1)**2)

# print the loss for the intial points: 
print("The total for the intial guess is ", function_f(w_1,w_2, b))
```

```python
# Next, we define the gradients. these will be used in our gradient descent algorithm.
def df_dw1(w_1, w_2, b): 
  return  4.2*w_1 + 6.3*w_2+0.1*b-4.1
def df_dw2(w_1, w_2, b): 
  return  6.6*w_1 + 36*w_2+4*b-16
def df_db(w_1, w_2, b): 
  return  0.2*w_1 + 4*w_2+8*b

  
```

```python
# Here, we define the number of iterations and the step size. Then, we train (learn).

number_of_iterations = 100
step_size            = 0.01

loss_values_list_to_save = []

# training loop: 
for training_iteration in range(number_of_iterations):

  # update w_1: 
  w_1 = w_1 - step_size * df_dw1(w_1, w_2, b)
  # update w_2: 
  w_2 = w_2 - step_size * df_dw2(w_1, w_2, b)
  # update b: 
  b = b - step_size * df_db(w_1, w_2, b)

  # update loss after getting new values for w1, w2, and b
  current_loss_value = function_f(w_1, w_2, b)

  loss_values_list_to_save.append(current_loss_value)

  # logger: 
  #print("Iteration = ", training_iteration, " ; Loss f = ", current_loss_value)
```

```python
# Next, we use the matplotlib to see the progress. First, we need to import the required libraries. 
# Also, we import the numpy library. 

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# define the axes
x_axis = list(range(number_of_iterations))
y_axis = loss_values_list_to_save
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```
